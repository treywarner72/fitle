```python
data = np.random.normal(size=5000000)
```

```python
model = fl.Param.positive('n') * fl.pdfs.gaussian()
```

```python
def number(n):
    return fl.fit(model | fl.Cost.chi2(data, bins=n, zero_method='absolute'), numba=False)
```

```python
number(18)
```
