```python
class Divider:
    """
    Divides parameter into n dimensions of categories
    """

    # converts 1 dimensional index to multidimensional
    @staticmethod
    def dimensionalise(dimensions, i):
        ret = []
        for dim in reversed(dimensions):
            ret.append(i%dim)
            i = math.floor(i/dim)
        ret.reverse()
        return tuple(ret)

    @staticmethod
    def split(param, n):
        split_vals = [np.percentile(param, i) for i in np.linspace(0,100,n+1)][1:-1]
        return [param] + split_vals
    
    # splits parameter by list of conditions.
    def __init__(self, parameter, cuts, conditions):
        # conditions
        self.parameter = parameter[cuts]
        self.cuts = cuts

        self.variables = []
        self.split_values = []
        self.conditions = []
        
        for cond in conditions:
            if type(cond[1]) != list:
                self.variables.append(cond[0][cuts])
                self.split_values.append(cond[1:])
                cond_split_vals = [-np.inf] + list(self.split_values[-1]) + [np.inf]
                self.split_values[-1] = cond_split_vals
                self.conditions.append([(self.variables[-1] >= cond_split_vals[i]) & (self.variables[-1] < cond_split_vals[i+1]) for i in range(len(cond_split_vals) -1)])
            else: 
                # boolean list, labels need to be added as a list to Divider.plot, not a string
                # automatic x for plotting will also not work, but not sure how you would plot boolean lists that you can't just define as splitted values
                self.conditions.append([c[cuts] for c in cond])
                self.split_values.append([])

        self.minimizers=[]
        self.dimensions = [len(dim) for dim in self.conditions]

    def _divide(self):
        self.param_conditions=[True]
        rev_split = list(reversed(self.conditions))

        for j in range(len(self.conditions)):
            dimension = rev_split[j]
            copy_conditions = np.copy(self.param_conditions[0])
            self.param_conditions[0]=[copy_conditions & cut for cut in dimension]

        #self.param_conditions = self.param_conditions[0]

    def fit(self, function, *args):
        self._divide()

        cond1d = np.array(self.param_conditions).reshape((np.prod(self.dimensions), len(self.parameter)))
        self.minimizers = np.empty(self.dimensions, dtype=object)
        
        for i in range(len(cond1d)):
            self.minimizers[Divider.dimensionalise(self.dimensions, i)] = function(self.parameter[cond1d[i]],*args)

    def labels1d(self, labels = [], label_format = "{lower:.2f} < {label} < {upper:.2f}"):
        ret_labels=[]
        for i in range(len(labels)):
            if type(labels[i]) == type("string"):
                labels[i] = [label_format.format(i=j, label=labels[i], lower=self.split_values[i][j], upper=self.split_values[i][j+1]) for j in range(len(self.split_values[i])-1)]
        for i in range(math.floor(np.prod(self.dimensions[:-1]))):
            label_indeces = Divider.dimensionalise(self.dimensions[:-1], i)
            ret_labels.append(", ".join([labels[label_i][label_indeces[label_i]] for label_i in range(len(label_indeces))]))
        return ret_labels

    def minimizer1d(self):
        return self.minimizers.reshape((np.prod(self.dimensions[:-1]), self.dimensions[-1])) if len(self.dimensions)>1 else [self.minimizers]


    def plot_vals(self, params, x=-1):
        minimizer1d = self.minimizer1d()
        values1d = np.empty_like(minimizer1d)
        errors1d = np.empty_like(minimizer1d)      

        for i in range(len(minimizer1d)):
            for j in range(len(minimizer1d[i])):
                values1d[i][j] = minimizer1d[i][j].values()
                errors1d[i][j] = minimizer1d[i][j].errors()
                for param in params:
                    values1d[i][j] = values1d[i][j][param]
                    errors1d[i][j] = errors1d[i][j][param]
        
        plot_x = []
        
        # x is from category variables, does not work for boolean list categories
        if type(x) == int:
            plot_x = np.array([[np.array(self.variables[x][cond]).mean() for cond in self.conditions[x]]] * len(minimizer1d))
        # x is from fit parameters, accessed the same way as params are for y
        elif type(x) == list or type(x) == tuple:
            plot_x = np.empty_like(minimizer1d)
            for i in range(len(minimizer1d)):
                for j in range(len(minimizer1d[i])):
                    plot_x[i][j] = minimizer1d[i][j].values()
                    for param in x:
                        plot_x[i][j] = plot_x[i][j][param]
                        
        return plot_x, values1d, errors1d
        
    
    # automatic plotting
    def plot(self, params, labels=[], label_format = "{lower:.2f} < {label} < {upper:.2f}",  format={}, x=-1):
        format = {'linestyle':'', 'marker':'.', 'ecolor':'black', **format}
        labels1d = self.labels1d(labels, label_format)
        plot_x, values1d, errors1d = self.plot_vals(params, x)

        for i in range(len(values1d)):
            plt.errorbar(plot_x[i],values1d[i], yerr=errors1d[i], label = labels1d[i], **format)#, color=colors[i])

        
        if len(self.split_values) > 1:
            plt.legend()

        return plot_x, values1d, errors1d
```

```python
def fit_data(data, range, bins):
    chi2 = fitlib.Fitter.binned(data, bins, range)
    chi2.mu = fitlib.f(1850,1900)
    chi2.pdf = [tanh_kpp.pdf(mu=chi2.mu), tanh_kpp.pdf(mu=chi2.mu), fitlib.Exp(x0=1790)]
    chi2.chi2()
    return chi2
```

```python
def fit_data_mle(data, range, bins):
    chi2 = fitlib.Fitter.binned(data, bins, range)
    chi2.mu = fitlib.f(1860,1880)
    chi2.w1 = fitlib.f(0,0.25,1)
    chi2.w2 = fitlib.f(0,0.25,1)
    n_tot = np.sum(chi2.y)
    chi2.pdf = [tanh_kpp.pdf(n=n_tot*chi2.w1, mu=chi2.mu), tanh_kpp.pdf(n=n_tot*chi2.w2*(1-chi2.w1),mu=chi2.mu), fitlib.Fit_function(lambda x,a,b: b*a/(np.e**(180*a)-1)*np.e**(a*(x-1790)), [fitlib.f(-np.inf,-0.01,0), n_tot*(1-chi2.w1-chi2.w2*(1-chi2.w1))]4)]
    chi2.BinnedMLE(999999)
    print(chi2.minimizer.valid)
    return chi2
```
