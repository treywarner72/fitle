```python
data = np.random.normal(size=1000000)
```

```python
model = fl.Param.positive('n') * fl.pdfs.gaussian()
```

```python
def number(n):
    return fl.fit(model | fl.Cost.chi2(data, bins=n, zero_method='absolute'), numba=False)
```

```python
fl.Cost.chi2(data, bins=100, zero_method='absolute').bin_widths
```
